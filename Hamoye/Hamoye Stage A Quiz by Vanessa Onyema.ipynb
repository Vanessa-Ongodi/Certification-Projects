```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
food_data=pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='ISO-8859-1')
food_data
```

```python
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively?
animal_fat_2014 = food_data.groupby('Item')['Y2014'].sum()
animal_fat_2014.to_frame()
```

```python
#total sum of animal fat in 2017
animal_fat_2017 = food_data.groupby('Item')['Y2017'].sum()
animal_fat_2017.to_frame()
```

```python
#What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
print(food_data['Y2015'].mean())
print(food_data['Y2015'].std())
```

```python
#What is the total number and percentage of missing data in 2016 to 2 decimal places?
total_null_2016 = food_data['Y2016'].isnull().sum()
print(total_null_2016)
percent_null_2016 = food_data['Y2016'].isnull().sum() * 100 / len(food_data)
print(percent_null_2016)
```

```python
#Which year had the highest correlation with â€˜Element Codeâ€™? 
food_data.corr().abs()
```

```python
#What year has the highest sum of Import Quantity? #for this I checked all the years and 2017 had the highest
sum_import_2017 = food_data.groupby('Element')['Y2017'].sum()
sum_import_2017.to_frame()
```

```python
#What is the total number of the sum of Production in 2014?
sum_2014 = food_data.groupby('Element')['Y2014'].sum()
sum_2014.to_frame()
```

```python
#Which of these elements had the highest sum in 2018? Which of these elements had the 3rd lowest sum in 2018?
sum_2018 = food_data.groupby('Element')['Y2018'].sum()
sum_2018.to_frame()
```
