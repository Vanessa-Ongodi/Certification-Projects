Amazon, the popular e-commerce platform is home to a lot of households and personal items ranging from electronics to books to fashion. The platform houses different books from different authors and every year most of these books make the best-selling list. This project focuses on the top 30 best-selling books from 2009 to 2019.

**Why webscrapping was used in this project?**

The amazon website contains unstructured data that can only be retrived by sending a request to the html and retrieving the data. Manually collecting this information could lead to loss of important information and also time so webscrapping the best selling books makes sure that the data accuracy is maintained and the metadata created will be trusted for reuse.

# Combine the code for all the books

```python
def extract_record(item):
    """Extract and return data from single record"""
    
    #description 
    atag =item.span
    description = atag.text.strip()
    title = item.find('div', '_p13n-zg-list-grid-desktop_truncationStyles_p13n-sc-css-line-clamp-1__1Fn1y').text
    author = item.find('div', 'a-row a-size-small').text
    
    #price
    price_parent = item.find('span', 'a-size-base a-color-price')
    price = price_parent.find('span', '_p13n-zg-list-grid-desktop_price_p13n-sc-price__3mJ9Z').text
    
    #rank and rating
    rating= item.i.text
    review_count = item.find('span', {'class': 'a-size-small'}).text
    
    #cover
    cover = item.find('span', 'a-size-small a-color-secondary a-text-normal').text
    
    result = (description, title , author, price, rating, review_count, cover)
    
    return result
   
```

# Export as csv

```python
records = []
results = soup.find_all('div', {'data-a-card-type': 'basic'})

for item in results:
    records.append(extract_record(item))
     #save data to csv file
    with open('results2009.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Description', 'Title', 'Author', 'Price', 'Rating', 'ReviewCount', 'Cover'])
        writer.writerows(records) 
```

To properly visualize and identify patterns with book buying, the exported csv files were edited in excel and a search was done for each of the books to identify the type of book and the genre. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import
Top_books_2009 = pd.read_csv('Top books 2009.csv')
Top_books_2010 = pd.read_csv('Top books 2010.csv')
Top_books_2011 = pd.read_csv('Top books 2011.csv')
Top_books_2012 = pd.read_csv('Top books 2012.csv')
Top_books_2013 = pd.read_csv('Top books 2013.csv')
Top_books_2014 = pd.read_csv('Top books 2014.csv')
Top_books_2015 = pd.read_csv('Top books 2015.csv')
Top_books_2016 = pd.read_csv('Top books 2016.csv')
Top_books_2017 = pd.read_csv('Top books 2017.csv')
Top_books_2018 = pd.read_csv('Top books 2018.csv')
Top_books_2019 = pd.read_csv('Top books 2019.csv')
```

## 2009

```python
Top_books_2009.head()
```

```python
#add a year column
Top_books_2009['Year'] = 2009
```

```python
#create the plot
plt.subplots(figsize=(10,10))
data=Top_books_2009["Book"]
data.value_counts().plot.pie(autopct='%1.1f%%',fontsize=20,colors=sns.color_palette('Set2',10),startangle=40)
plt.title("Type of Book", fontsize=25)
plt.ylabel('')
plt.savefig("book2009.svg")
plt.show()
```

```python
#genre of fiction
fiction_2009 = Top_books_2009.loc[Top_books_2009['Book'].isin(["Fiction"])]
fiction_type_2009=fiction_2009['Genre'].value_counts()
fiction_type_2009 = fiction_type_2009.to_frame().reset_index()
fiction_type_2009
```

```python
#create a bar chart using Seaborn
sns.set_style("whitegrid")
bar,ax = plt.subplots(figsize=(5,5))
ax = sns.barplot(x='Genre', y='index', data=fiction_type_2009, ci=None, color='grey',orient='h')
ax.set_title('Genre of Fiction', fontsize=25)
ax.set_xlabel ("count", fontsize=20)
ax.set_ylabel (None, fontsize=20)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
plt.savefig("fiction2009.svg")
```

```python
nonfiction_2009 = Top_books_2009.loc[Top_books_2009['Book'].isin(["Non-Fiction"])]
nonfiction_type_2009=nonfiction_2009['Genre'].value_counts()
nonfiction_type_2009 = nonfiction_type_2009.to_frame().reset_index()
nonfiction_type_2009
```
